```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

import joblib
```

```python
df = pd.read_csv("data\CropYield\yield_df.csv")

df.head()
```

```python
print("Shape :", df.shape)

print("\nColumns")

print(df.columns)

print("\nMissing Values")

print(df.isnull().sum())
```

```python
# Remove index column
df.drop("Unnamed: 0", axis=1, inplace=True)

# Verify
print(df.head())
print("\nShape :", df.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create encoders
area_encoder = LabelEncoder()
item_encoder = LabelEncoder()

# Encode categorical columns
df["Area"] = area_encoder.fit_transform(df["Area"])
df["Item"] = item_encoder.fit_transform(df["Item"])

print(df.head())
```

```python
print(df.dtypes)
```

```python
import joblib

joblib.dump(area_encoder, "area_encoder.pkl")
joblib.dump(item_encoder, "item_encoder.pkl")

print("Encoders Saved Successfully")
```

```python
# Features
X = df.drop("hg/ha_yield", axis=1)

# Target
y = df["hg/ha_yield"]

print("Feature Shape :", X.shape)
print("Target Shape :", y.shape)

print("\nFeatures:")
print(X.columns)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Features :", X_train.shape)
print("Testing Features :", X_test.shape)

print("Training Target :", y_train.shape)
print("Testing Target :", y_test.shape)
```

```python
print(X_train.head())
```

```python
from sklearn.linear_model import LinearRegression

# Create model
lr_model = LinearRegression()

# Train model
lr_model.fit(X_train, y_train)

print("Linear Regression Model Trained Successfully")
```

```python
lr_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

lr_mae = mean_absolute_error(y_test, lr_pred)
lr_rmse = np.sqrt(mean_squared_error(y_test, lr_pred))
lr_r2 = r2_score(y_test, lr_pred)

print("Linear Regression Performance")
print("-"*35)

print("MAE :", lr_mae)
print("RMSE:", lr_rmse)
print("R2 Score:", lr_r2)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=42)

dt_model.fit(X_train, y_train)

print("Decision Tree Model Trained Successfully")
```

```python
dt_pred = dt_model.predict(X_test)
```

```python
dt_mae = mean_absolute_error(y_test, dt_pred)
dt_rmse = np.sqrt(mean_squared_error(y_test, dt_pred))
dt_r2 = r2_score(y_test, dt_pred)

print("Decision Tree Performance")
print("-"*35)

print("MAE :", dt_mae)
print("RMSE:", dt_rmse)
print("R2 Score:", dt_r2)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)

print("Random Forest Model Trained Successfully")
```

```python
rf_pred = rf_model.predict(X_test)
```

```python
rf_mae = mean_absolute_error(y_test, rf_pred)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
rf_r2 = r2_score(y_test, rf_pred)

print("Random Forest Performance")
print("-"*35)

print("MAE :", rf_mae)
print("RMSE:", rf_rmse)
print("R2 Score:", rf_r2)
```

```python
results = pd.DataFrame({
    "Model": [
        "Linear Regression",
        "Decision Tree",
        "Random Forest"
    ],
    "MAE": [
        lr_mae,
        dt_mae,
        rf_mae
    ],
    "RMSE": [
        lr_rmse,
        dt_rmse,
        rf_rmse
    ],
    "R2 Score": [
        lr_r2,
        dt_r2,
        rf_r2
    ]
})

results = results.sort_values(by="R2 Score", ascending=False)

results
```

```python
import joblib

joblib.dump(rf_model, "yield_prediction_model.pkl")

print("Model Saved Successfully")
```

## Summary

1. The crop yield dataset was preprocessed by removing unnecessary columns and encoding categorical features (Area and Item) using Label Encoding.

2. The dataset was divided into training (80%) and testing (20%) sets for model development and evaluation.

3. Three regression models—Linear Regression, Decision Tree Regressor, and Random Forest Regressor—were trained to predict crop yield.

4. Model performance was evaluated using Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), and R² Score.

5. Random Forest achieved the highest prediction accuracy with the lowest error values, outperforming the other regression models.

6. The trained Random Forest model was selected as the final crop yield prediction model.

7. The final model was saved for future predictions and deployment.
