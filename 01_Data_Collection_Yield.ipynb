```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
yield_df = pd.read_csv(r"data\CropYield\yield_df.csv")
rainfall = pd.read_csv(r"data\CropYield\rainfall.csv")
temperature = pd.read_csv(r"data\CropYield\temp.csv")
pesticides = pd.read_csv(r"data\CropYield\pesticides.csv")
yield_data = pd.read_csv(r"data\CropYield\yield.csv")
```

```python
datasets = {
    "yield_df": yield_df,
    "rainfall": rainfall,
    "temperature": temperature,
    "pesticides": pesticides,
    "yield": yield_data
}

for name, df in datasets.items():
    print(f"\n{name}")
    print("-"*30)
    print("Rows :", df.shape[0])
    print("Columns :", df.shape[1])
```
